```python
def set_target_value(dataset, values):
    class_values = [row[values] for row in dataset]
    distinct = set(class_values)
    set_int = {}
    for i, value in enumerate(distinct):
        set_int[value] = i
    for row in dataset:
        row[values] = set_int[row[values]]
    return set_int


set_target_value(dataset, len(dataset[0])-1)
```

```python
# Using cross validation method to split into different training and testing sets each time and avoids biasing

def cross_val(dataset,k):
    dataset1=list(dataset)    # Making a copy of it to split into 5 parts here which would vary each time
    setn_data=[]              # Each validation returns a different set of training data each time to avoid biasing
    no_of_parts=int(len(dataset)/k)           # 5 fold here #87
    for i in range(k):
        part=[]
        while len(part) < no_of_parts:
            index=randrange(len(dataset1))
            part.append(dataset1.pop(index))
        setn_data.append(part)
    return setn_data               #Returning the split data
```

```python
# How accurate is the implemented model

def accuracy_check(actual,predicted):
    total_crt = 0                               # How many values are predicted correctly with respect to actual
    for i in range(len(actual)):
        if actual[i] == predicted[i]:
            total_crt+=1
    percentage = total_crt / float(len(actual)) * 100.0          # Calculating accuracy percentage
    return percentage
```

```python
#Model designed with reference from the references mentioned
def seperateByClass(dataset):
    seperated={}
    for i in range(len(dataset)):
        v = dataset[i]
        if (v[-1] not in seperated):
            seperated[v[-1]]=list()
        seperated[v[-1]].append(v)
    return seperated
    
def mean(values):
    average=sum(values)/float(len(values))
    return average

def stdev(values):
    avg = mean(values)
    variance = sum([(x-avg)**2 for x in values])/float(len(values)-1)
    return sqrt(variance)
    
def statistics(dataset):
    stats = [(mean(each_col),stdev(each_col),len(each_col)) for each_col in zip(*dataset)]
    del(stats[-1])
    return stats

def value_targetClass(dataset):
    stat={}
    seperated=seperateByClass(dataset)
    for target,rows in seperated.items():
        stat[target]=statistics(rows)
    return stat


def cal_probablity(x, mean, stdev):
    numerator=(x-mean)**2
    denominator=2 * stdev**2
    exponent=exp(-(numerator/denominator))
    return (1 / (sqrt(2 * pi) * stdev)) * exponent


def prob_by_class(stats,rows):
    total_rows=sum([stats[class_value][0][2] for class_value in stats])
    probablity={}
    for target,summary in stats.items():
        probablity[target]=stats[target][0][2]/float(total_rows)
        for i in range(len(summary)):
            mean, stdev, _ = summary[i]
            probablity[target] *= cal_probablity(rows[i],mean,stdev)
    return probablity

def predict_value(stats,rows):
    probablity = prob_by_class(stats,rows)
    best_class,best_prob=None,-1
    for target,pb in probablity.items():
        if pb > best_prob or best_class is None:
            best_prob = pb
            best_class = target
    return best_class
    
    
# The classification function with respect to the training and the testing sets obtained for this run
def naive_bayes_classifier(trainingSet,testingSet):
    prediction = []
    value = value_targetClass(trainingSet)
    for i in testingSet:
        p = predict_value(value,i)
        prediction.append(p)
    return(prediction)
```

```python
def confusion_matrix(actual,pred):
    TP=0
    FP=0
    TN=0
    FN=0
    
    for i in range(len(pred)):
        if actual[i]==pred[i]==1:
            TP+=1
        if pred[i]==1 and actual[i]!=pred[i]:
            FP+=1
        if actual[i]==pred[i]==0:
            TN+=1
        if pred[i]==0 and actual[i]!=pred[i]:
            FN+=1
    return (TP,FP,TN,FN)

def naive_model(dataset,classifier,k,*args):
    accuracies=[]
    parts = cross_val(dataset,k)
    
    for part in parts: #setn_data
        
        train=list(parts)
        train.remove(part)
        train=sum(train,[])
        test=[]
        
        for row in part:
            row1=list(row)
            test.append(row1)
            row1[-1]=None
            
        predicted_value=classifier(train,test,*args)
        actual_value=[row[-1] for row in part]
        
        accuracy=accuracy_check(actual_value,predicted_value)
        TP,FP,TN,FN=confusion_matrix(actual_value,predicted_value)
        rows,cols=(2,2)
        conf_mat=[[0 for j in range(cols)] for i in range(rows)]
        conf_mat[0][0]=TP
        conf_mat[0][1]=FP
        conf_mat[1][0]=FN
        conf_mat[1][1]=TN
        print("Confusion matrix:")
        for i in range(rows):
            for j in range(cols):
                print(conf_mat[i][j],end=" ")
            print()
        print('\n')

        accuracies.append(accuracy)
        
    return (accuracies,TP,FP,TN,FN)
```
